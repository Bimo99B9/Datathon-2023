```python
df_result = audiencia.join(movilidad)

display(df_result)
```

```python
# Select the columns for the weather variables and the TV audience variables
tv_audience_columns = ['MAud_TTV_4plus', 'AM_TTV_4plus']
movilidad_columns = ['retail_and_recreation_percent_change_from_baseline',
       'grocery_and_pharmacy_percent_change_from_baseline',
       'parks_percent_change_from_baseline',
       'transit_stations_percent_change_from_baseline',
       'workplaces_percent_change_from_baseline',
       'residential_percent_change_from_baseline']

# Calculate the correlation between the selected columns
correlation_matrix = df_result[tv_audience_columns + movilidad_columns].corr()

# Show the correlation between the weather variables and the TV audience variables
correlation_result = correlation_matrix.loc[movilidad_columns,tv_audience_columns]

# Create a heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_result, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1)

# Set the title and labels
plt.title('Correlación entre Variables de Movilidad y TV Audiencia Variables')
plt.xlabel('TV Audiencia Variables')
plt.ylabel('Movilidad Variables')

# Show the plot
plt.show()
```
